- ### **Cambia el formato de las aotaciones**

```python
file_path   = './augmented/annotated dataset voc'
files_train = glob(os.path.join(file_path, 'train', '*.xml'))
files_test  = glob(os.path.join(file_path, 'test' , '*.xml'))
files       = files_train + files_test

for file in files:
    tree = ET.parse(file)
    root = tree.getroot()

    data = {
        'filename': root.find('filename').text,
        'image_width': int(root.find('./size/width').text),
        'image_height': int(root.find('./size/height').text),
        'name'       : [obj.find('name').text for obj in root.findall('object')],
        'xmin'       : [int(obj.find('./bndbox/xmin').text) for obj in root.findall('object')],
        'ymin'       : [int(obj.find('./bndbox/ymin').text) for obj in root.findall('object')],
        'xmax'       : [int(obj.find('./bndbox/xmax').text) for obj in root.findall('object')],
        'ymax'       : [int(obj.find('./bndbox/ymax').text) for obj in root.findall('object')]
    }

    # Create a DataFrame from the dictionary
    df = pd.DataFrame(data)

    df ['x_center'] = (df['xmin'] + df['xmax']) / 2 / df['image_width']
    df ['y_center'] = (df['ymin'] + df['ymax']) / 2 / df['image_height']
    df ['width']    = (df['xmax'] - df['xmin']) / df['image_width']
    df ['height']   = (df['ymax'] - df['ymin']) / df['image_height']

    df = df[['name', 'x_center', 'y_center', 'width', 'height']]
    df = df.rename(columns={'name':'class_index'})
    df['class_index'] = np.where(df['class_index'] == 'Head', 1, 0)
    
    out_path = "".join((os.path.splitext(file)[0], '.txt'))
    df.to_csv(out_path, sep=' ', header=False, index=False)
    
    print(out_path)
```

- ### **Divide el dataset tanto los paths, como las imágenes y las etiquetas**
    - Dataset de entrenamiento 80%
    - Dataset de test 15%
    - Dataset de validación 5%

- **Lista todas las imagenes**

```python
def open_dataframe(path):
    try:
        return pd.read_csv(path, header=None)
    except pd.errors.EmptyDataError:
        return pd.DataFrame()
```

```python
def modify_path_less(path):
    parts = path.split('/')  # Split the path by '/'
    parts.pop(2)  # Remove the 'results' directory
    return '/'.join(parts)  # Join the path back together
```

```python
path_train = './augmented/annotated ultralytics/train.txt'
path_val   = './augmented/annotated ultralytics/val.txt'

df_train = open_dataframe(path_train)
df_val   = open_dataframe(path_val)
    
df = pd.concat([df_train, df_val])
df.columns = ['Paths']
df['Paths'] = df['Paths'].apply(modify_path_less)

string_to_replace = './images'
replacement_string = './augmented/annotated ultralytics/images'
df['Paths_images'] = df.Paths.str.replace(string_to_replace, replacement_string)

replacement_string = './augmented/annotated ultralytics/labels'
df['Paths_labels'] = df.Paths.str.replace(string_to_replace, replacement_string)
df['Paths_labels'] = df.Paths_labels.str.replace('.jpg$', '.txt', regex=True)

df = df.drop_duplicates()
```

- **Separa el dataset**

```python
train_set, temp = train_test_split(df, test_size=0.2, random_state=42)

# Splitting the remaining 20% into two sets: 15% and 5%
val_set, test_set = train_test_split(temp, test_size=0.75, random_state=42)
```

- **Exporta los diferentes archivos**

```python
val_set
```

```python
train_set = train_set.drop_duplicates()
val_set   = val_set.drop_duplicates()
test_set  = test_set.drop_duplicates()

train_set['Paths'] = train_set['Paths'].str.replace('./images', './images/train')
val_set  ['Paths'] = val_set  ['Paths'].str.replace('./images', './images/val')
test_set ['Paths'] = test_set ['Paths'].str.replace('./images', './images/test')


train_set.Paths.to_csv('train_80.txt', quoting=0, quotechar=" ", header=False, index=False, sep='/', mode='w')
val_set.Paths.to_csv  ('val_15.txt'  , quoting=0, quotechar=" ", header=False, index=False, sep='/', mode='w')
test_set.Paths.to_csv ('test_05.txt' , quoting=0, quotechar=" ", header=False, index=False, sep='/', mode='w')
```

- **Mueve los archivos a las diferentes carpetas**

```python
def create_new_path(path):
    if os.path.isdir(path) == False:
        os.makedirs(path)
```

```python
out_path_image = os.path.dirname(train_set.iloc[0].Paths_images)

# Mueve imagenes a las nuevas rutas
out_path_image_train = os.path.normpath(os.path.join(out_path_image, 'train'))
create_new_path(out_path_image_train)
[shutil.move(file, out_path_image_train) for file in train_set.Paths_images.values]

out_path_image_val = os.path.normpath(os.path.join(out_path_image, 'val'))
create_new_path(out_path_image_val)
[shutil.move(file, out_path_image_val) for file in val_set.Paths_images.values]

out_path_image_test = os.path.normpath(os.path.join(out_path_image, 'test'))
create_new_path(out_path_image_test)
[shutil.move(file, out_path_image_test) for file in test_set.Paths_images.values]
```

```python
out_path_label = os.path.dirname(train_set.iloc[0].Paths_labels)

# Mueve los labels a las nuevas rutas
out_path_label_train = os.path.normpath(os.path.join(out_path_label, 'train'))
create_new_path(out_path_label_train)
[shutil.move(file, out_path_label_train) for file in train_set.Paths_labels.values]

out_path_label_val = os.path.normpath(os.path.join(out_path_label, 'val'))
create_new_path(out_path_label_val)
[shutil.move(file, out_path_label_val) for file in val_set.Paths_labels.values]

out_path_label_test = os.path.normpath(os.path.join(out_path_label, 'test'))
create_new_path(out_path_label_test)
[shutil.move(file, out_path_label_test) for file in test_set.Paths_labels.values]
```
